```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from soccerplots.radar_chart import Radar
from soccerplots.utils import add_image
```

```python
def prepare_data_CB(file1,file2,file3):
    data1 = pd.read_excel(file1)
    data2 = pd.read_excel(file2)
    data3 = pd.read_excel(file3)
    
    df1 = data1[['TotalCmp%','xA','Prog']]
    df2 = data2[['Fls','Won%','Won']]
    df3 = data3[['Pressures%','TacklesTkl','Int','Clr','BlocksBlocks','Err']]
    
    return pd.concat([df1,df2,df3],axis=1)


def plot_CB_radar(data,PlayerName,Club,Position,Season):
    
    r_data={}
    r_data['Total Pass Comp% ']=[data['TotalCmp%'].mean()]
    r_data['xA']=[data['xA'].mean()]
    r_data['Progressive Passes']=[data['Prog'].mean()]
    r_data['Fouls per game']=[data['Fls'].mean()]
    r_data['Aerials Won%']=[data['Won%'].mean()]
    r_data['Aerials Won']=[data['Won'].mean()]
    r_data['Pressures%']=[data['Pressures%'].mean()]
    r_data['Tackles']=[data['TacklesTkl'].mean()]
    r_data['Interception/game']=[data['Int'].mean()]
    r_data['Clearence/game']=[data['Clr'].mean()]
    r_data['Blocks/game']=[data['BlocksBlocks'].mean()]
    r_data['Error led to goal']=[data['Err'].mean()]    

    
#     return pd.DataFrame.from_dict(r_data)    
    plot_df = pd.DataFrame.from_dict(r_data)    
    params = plot_df.columns.tolist()
## range values
    ranges = [(82, 100), (0.009, 0.02), (1,5 ), (0, 0.8),
              (40,80),(1,5), (30,50), (0.1,2),
              (0.5,1.2),(2,5), (0.45,1.5), (.04,0.1)]

## parameter value
    values = plot_df.iloc[0]

## title values
    title = dict(
        title_name=PlayerName,
        title_color="#B6282F",
        subtitle_name=Club,
        subtitle_color="#B6282F",
        title_name_2=Season,
        title_color_2="#E3DDED",
        subtitle_name_2=Position,
        subtitle_color_2='#B6282F',
        title_fontsize=18,
        subtitle_fontsize=20)


## instantiate object
    radar = Radar()
#     radar = Radar(background_color="#121212", patch_color="#28252C", label_color="#FFFFFF",
#               range_color="#FFFFFF")
## plot radar
    fig, ax = radar.plot_radar(ranges=ranges, params=params, 
                           values=values, radar_color=['#1c547f', '#CF6679'], 
                           title=title)
    fig = add_image(image="thaigo.jpg", fig=fig, left=0.464, bottom=0.81, width=0.1, height=0.075)
    
def plot_compare_radar_CB(df1,df2):

    r_df1={}
    r_df1['Total Pass Comp% ']=[df1['TotalCmp%'].mean()]
    r_df1['xA']=[df1['xA'].mean()]
    r_df1['Progressive Passes']=[df1['Prog'].mean()]
    r_df1['Fouls per game']=[df1['Fls'].mean()]
    r_df1['Aerials Won%']=[df1['Won%'].mean()]
    r_df1['Aerials Won']=[df1['Won'].mean()]
    r_df1['Pressures%']=[df1['Pressures%'].mean()]
    r_df1['Tackles']=[df1['TacklesTkl'].mean()]
    r_df1['Interception/game']=[df1['Int'].mean()]
    r_df1['Clearence/game']=[df1['Clr'].mean()]
    r_df1['Blocks/game']=[df1['BlocksBlocks'].mean()]
    r_df1['Error led to goal']=[df1['Err'].mean()]    


    r_df2={}
    r_df2['Total Pass Comp% ']=[df2['TotalCmp%'].mean()]
    r_df2['xA']=[df2['xA'].mean()]
    r_df2['Progressive Passes']=[df2['Prog'].mean()]
    r_df2['Fouls per game']=[df2['Fls'].mean()]
    r_df2['Aerials Won%']=[df2['Won%'].mean()]
    r_df2['Aerials Won']=[df2['Won'].mean()]
    r_df2['Pressures%']=[df2['Pressures%'].mean()]
    r_df2['Tackles']=[df2['TacklesTkl'].mean()]
    r_df2['Interception/game']=[df2['Int'].mean()]
    r_df2['Clearence/game']=[df2['Clr'].mean()]
    r_df2['Blocks/game']=[df2['BlocksBlocks'].mean()]
    r_df2['Error led to goal']=[df2['Err'].mean()]   
    
    
    plot_df = pd.DataFrame.from_dict(r_df1)    
    plot_df = plot_df.append(pd.DataFrame.from_dict(r_df2), ignore_index=True)
    
#     return plot_df
    params = plot_df.columns.tolist()


## range values
    ranges = [(82, 100), (0.009, 0.02), (1,5 ), (0, 0.8),
              (40,80),(1,5), (30,50), (0.1,2),
              (0.5,1.2),(2,5), (0.45,1.5), (.04,0.1)]

## parameter value
    values = [plot_df.iloc[0],plot_df.iloc[1]]

## title
    title = dict(
        title_name='Thiago Silva',
        title_color='#B6282F',
        subtitle_name='Chelsea(2020-21)',
        subtitle_color='#B6282F',
        title_name_2='Thiago Silva',
        title_color_2='#344D94',
        subtitle_name_2='PSG(2019-20)',
        subtitle_color_2='#344D94',
        title_fontsize=18,
        subtitle_fontsize=15,
    )


## instantiate object
    radar = Radar()

## plot radar -- compare
    fig, ax = radar.plot_radar(ranges=ranges, params=params, values=values, 
                           radar_color=['#B6282F', '#344D94'], 
                           title=title,compare=True)
    fig = add_image(image="thaigo.jpg", fig=fig, left=0.464, bottom=0.81, width=0.1, height=0.075)

    
def prepare_data_FB(file1,file2,file3):
    data1 = pd.read_excel(file1)
    data2 = pd.read_excel(file2)
    data3 = pd.read_excel(file3)
    
    df1 = data1[['TotalCmp%','xA','Prog','CrsPA']]
    df2 = data2[['Fls','Won%']]
    df3 = data3[['Pressures%','TacklesTkl','Int','Clr','BlocksBlocks','DribblesTkl%']]
    
    return pd.concat([df1,df2,df3],axis=1)

def plot_FB_radar(data,PlayerName,Club,Position):
    
    r_data={}
    r_data['Total Pass Comp% ']=[data['TotalCmp%'].mean()]
    r_data['xA']=[data['xA'].mean()]
    r_data['Progressive Passes']=[data['Prog'].mean()]
    r_data['Crosses into 18-yard']=[data['CrsPA'].mean()]    
    r_data['Fouls per game']=[data['Fls'].mean()]
    r_data['Aerials Won%']=[data['Won%'].mean()]
    r_data['Pressures%']=[data['Pressures%'].mean()]
    r_data['Tackles/game']=[data['TacklesTkl'].mean()]
    r_data['Interception/game']=[data['Int'].mean()]
    r_data['Clearence/game']=[data['Clr'].mean()]
    r_data['Blocks/game']=[data['BlocksBlocks'].mean()]
    r_data['% Dribblers tackled']=[data['DribblesTkl%'].mean()]    

    
#     return pd.DataFrame.from_dict(r_data)    
    plot_df = pd.DataFrame.from_dict(r_data)    
    params = plot_df.columns.tolist()
## range values
    ranges = [(50,80), (0.09, 0.18), (1,5 ), (0, 0.8),
              (0.4, 1.8),(40,80), (20,30), (1,3),
              (0.5,1),(2,5), (0.45,2.5), (30,45)]

## parameter value
    values = plot_df.iloc[0]

## title values
    title = dict(
        title_name=PlayerName,
        title_color="#B6282F",
        subtitle_name=Club,
        subtitle_color="#B6282F",
        title_name_2="2020-21",
        title_color_2="#E3DDED",
        subtitle_name_2=Position,
        subtitle_color_2='#B6282F',
        title_fontsize=18,
        subtitle_fontsize=20)


## instantiate object
    radar = Radar()
#     radar = Radar(background_color="#121212", patch_color="#28252C", label_color="#FFFFFF",
#               range_color="#FFFFFF")
## plot radar
    fig, ax = radar.plot_radar(ranges=ranges, params=params, 
                           values=values, radar_color=['#1c547f', '#CF6679'], 
                           title=title)
    fig = add_image(image="ben.jpg", fig=fig, left=0.464, bottom=0.81, width=0.1, height=0.075)
```

```python
plot_prem_CB = prepare_data_CB('thiago_prem_pass.xlsx','thiago_prem_misc.xlsx','thiago_prem_def.xlsx')
plot_CB_radar(plot_prem_CB,'Thiago Silva','Chelsea','Defender','2020-21')
```
